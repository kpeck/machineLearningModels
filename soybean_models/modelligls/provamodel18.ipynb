```python
#importazione dei dati ------------------------------------------------------------------------------------------------------------

from pandas import DataFrame
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.linalg import toeplitz


df = pd.read_csv (r'/Users/patrickpinta/Desktop/1datasetpanda/soybean.csv',header=0,error_bad_lines=False) 

#modellazione dei dati training set------------------------------------------------------------------------------------------------------------


xExpUsa= df.iloc[0:250,1:2]
xExpUsa = np.array(xExpUsa)
xExpBra=df.iloc[0:250,3:4]
xExpBra = np.array(xExpBra)
xExpArg=df.iloc[0:250,5:6]
xExpArg = np.array(xExpArg)
xImportChina=xExpUsa+xExpBra+xExpArg

xBimestri=df.iloc[0:250,17:22]
xBimestri = np.array(xBimestri)

xUsa=df.iloc[0:250,0:1]
xUsa = np.array(xUsa)
xBimInfExpUsa=xBimestri*xUsa

xBra=df.iloc[0:250,1:2]
xBra = np.array(xBra)
xBimInfExpBra=xBimestri*xBra

xArg=df.iloc[0:250,2:3]
xArg = np.array(xArg)
xBimInfExpArg=xBimestri*xArg


x=df.iloc[0:250,0:1]
x2=df.iloc[0:250,2:3]
x3=df.iloc[0:250,4:5]
x4=df.iloc[0:250,6:7]
x5=df.iloc[0:250,8:15]
x6=df.iloc[0:250,15:17]
x7=df.iloc[0:250,17:22]


xImportChina=xImportChina.tolist()
xImportChina = DataFrame (xImportChina,columns=['ImportChina'])
xBimInfExpUsa=xBimInfExpUsa.tolist()
xBimInfExpUsa = DataFrame (xBimInfExpUsa,columns=['bim1*expUsa','bim2*expUsa','bim3*expUsa','bim4*expUsa','bim5*expUsa'])
xBimInfExpBra=xBimInfExpBra.tolist()
xBimInfExpBra = DataFrame (xBimInfExpBra,columns=['bim1*expBra','bim2*expBra','bim3*expBra','bim4*expBra','bim5*expBra'])
xBimInfExpArg=xBimInfExpArg.tolist()
xBimInfExpArg = DataFrame (xBimInfExpArg,columns=['bim1*expArg','bim2*expArg','bim3*expArg','bim4*expArg','bim5*expArg'])
x=x.join(xImportChina)
x=x.join(x2)
x=x.join(x3)
x=x.join(x4)
x=x.join(x5)
x=x.join(x6)
x=x.join(xBimInfExpUsa)
x=x.join(xBimInfExpBra)
x=x.join(xBimInfExpArg)
x=x.join(x7)


y= df.iloc[0:250,23:24]

#modellazione dei dati crossvalidation set------------------------------------------------------------------------------------------------------------

xExpUsacv= df.iloc[0:334,1:2]
xExpUsacv = np.array(xExpUsacv)
xExpBracv=df.iloc[0:334,3:4]
xExpBracv = np.array(xExpBracv)
xExpArgcv=df.iloc[0:334,5:6]
xExpArgcv = np.array(xExpArgcv)
xImportChinacv=xExpUsacv+xExpBracv+xExpArgcv

xBimestricv=df.iloc[0:334,17:22]
xBimestricv = np.array(xBimestricv)

xUsacv=df.iloc[0:334,0:1]
xUsacv = np.array(xUsacv)
xBimInfExpUsacv=xBimestricv*xUsacv

xBracv=df.iloc[0:334,1:2]
xBracv = np.array(xBracv)
xBimInfExpBracv=xBimestricv*xBracv

xArgcv=df.iloc[0:334,2:3]
xArgcv = np.array(xArgcv)
xBimInfExpArgcv=xBimestricv*xArgcv


xcv=df.iloc[250:334,0:1]
x2cv=df.iloc[250:334,2:3]
x3cv=df.iloc[250:334,4:5]
x4cv=df.iloc[250:334,6:7]
x5cv=df.iloc[250:334,8:15]
x6cv=df.iloc[250:334,15:17]
x7cv=df.iloc[250:334,17:22]


xImportChinacv=xImportChinacv.tolist()
xImportChinacv = DataFrame (xImportChinacv,columns=['ImportChina'])
xBimInfExpUsacv=xBimInfExpUsacv.tolist()
xBimInfExpUsacv = DataFrame (xBimInfExpUsacv,columns=['bim1*expUsa','bim2*expUsa','bim3*expUsa','bim4*expUsa','bim5*expUsa'])
xBimInfExpBracv=xBimInfExpBracv.tolist()
xBimInfExpBracv = DataFrame (xBimInfExpBracv,columns=['bim1*expBra','bim2*expBra','bim3*expBra','bim4*expBra','bim5*expBra'])
xBimInfExpArgcv=xBimInfExpArgcv.tolist()
xBimInfExpArgcv = DataFrame (xBimInfExpArgcv,columns=['bim1*expArg','bim2*expArg','bim3*expArg','bim4*expArg','bim5*expArg'])
xcv=xcv.join(xImportChinacv.iloc[250:334,0:1])
xcv=xcv.join(x2cv)
xcv=xcv.join(x3cv)
xcv=xcv.join(x4cv)
xcv=xcv.join(x5cv)
xcv=xcv.join(x6cv)
xcv=xcv.join(xBimInfExpUsacv.iloc[250:334,0:5])
xcv=xcv.join(xBimInfExpBracv.iloc[250:334,0:5])
xcv=xcv.join(xBimInfExpArgcv.iloc[250:334,0:5])
xcv=xcv.join(x7cv)

ycv= df.iloc[250:334,23:24]

#stardadizzazione dei dati training set------------------------------------------------------------------------------------------------------------

scaler = MinMaxScaler(feature_range=(-1,1))

x_train_stand=(scaler.fit_transform(x))

#stardadizzazione dei dati crossvalidation set------------------------------------------------------------------------------------------------------------

scaler = MinMaxScaler(feature_range=(-1,1))

xcv_train_stand=(scaler.fit_transform(xcv))
    
#modello dati training set------------------------------------------------------------------------------------------------------------    
    
x_train_stand, y = np.array(x_train_stand), np.array(y)
x_train_stand = sm.add_constant(x_train_stand)

#modello dati training set------------------------------------------------------------------------------------------------------------    

xcv_train_stand, ycv = np.array(xcv_train_stand), np.array(ycv)
xcv_train_stand = sm.add_constant(xcv_train_stand)

#procedura per sigma e per gls model ------------------------------------------------------------------------------------------------------------

ols_resid = sm.OLS(y.astype(float), x_train_stand.astype(float)).fit().resid
resid_fit = sm.OLS(ols_resid[1:], sm.add_constant(ols_resid[:-1])).fit()
rho = resid_fit.params[1]
toeplitz(range(5))
order = toeplitz(range(len(ols_resid)))
sigma = rho**order

gls_model = sm.GLS(y, x_train_stand,sigma)
gls_results = gls_model.fit()

#print(gls_results.summary())

#procedura per ols model ------------------------------------------------------------------------------------------------------------

ols_model = sm.OLS(y, x_train_stand)
ols_results = ols_model.fit(cov_type='HC1') 

print(gls_results.summary())

#identify outliers ------------------------------------------------------------------------------------------------------------
"""
import numpy as np
np.set_printoptions(suppress=True)

influence = ols_results.get_influence()

cooks = influence.cooks_distance

infl = influence.summary_frame().dfb_x1 > 2./250**.5

print(influence.summary_frame()[infl].filter(regex="dfb_x1"))

print(2./417**.5)

"""
#forecast ------------------------------------------------------------------------------------------------------------

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

ypredTrain = gls_results.predict(x_train_stand)
ycvpred = gls_results.predict(xcv_train_stand)

print(r2_score(y,ypredTrain))
print(r2_score(ycv,ycvpred))
print(mean_squared_error(y,ypredTrain))
print(mean_squared_error(ycv,ycvpred))







```
