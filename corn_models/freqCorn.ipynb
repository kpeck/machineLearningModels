```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df = pd.read_csv (r'/Users/patrickpinta/Desktop/1datasetpanda/corn.csv',header=0,error_bad_lines=False) 

x= df.iloc[0:334,18:19]
x = np.array(x)


# histogram and kernel density estimation function of the variable height
ax = sns.distplot(x, hist=True, hist_kws={"edgecolor": 'w', "linewidth": 3}, kde_kws={"linewidth": 3})


# ticks 
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# labels and title
plt.xlabel('prec price', fontsize=14)
plt.ylabel('frequency', fontsize=14)
plt.title('Distribution of prec price', fontsize=20);

plt.savefig('/Users/patrickpinta/Desktop/freqcorn/14h_prec_price.png')
```
